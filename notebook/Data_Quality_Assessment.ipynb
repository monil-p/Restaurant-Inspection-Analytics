# **Data Overview**

```python
#Load csv

import pandas as pd
import numpy as np

CSV_PATH = '/DOHMH_New_York_City_Restaurant_Inspection_Results_20260603.csv'

df = pd.read_csv(CSV_PATH)
```

```python
df.shape
df.head()
df.info()
```

# **Missing Values**

```python
missing = (
    df.isnull()
    .sum()
    .rename("null_count")
    .to_frame()
    .assign(pct=lambda x: (x["null_count"] / len(df) * 100).round(2))
    .sort_values("null_count", ascending=False)
)
print(missing[missing["null_count"] > 0].to_string() or "  No missing values.")
```

# **Duplicates**

```python
n_dup = df.duplicated().sum()
print(f"Fully duplicate rows: {n_dup:,}  ({n_dup / len(df):.1%})")
```

# **Categories**

```python
cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
if cat_cols:
    print("\n" + "=" * 60)
    print("CATEGORICAL COLUMNS")
    print("=" * 60)
    for col in cat_cols:
        n_unique = df[col].nunique()
        top = df[col].value_counts().head(5)
        print(f"\n  [{col}]  unique={n_unique:,}")
        for val, cnt in top.items():
            print(f"    {str(val):<30} {cnt:>6,}  ({cnt/len(df):.1%})")
        if n_unique > 5:
            print(f"    ... ({n_unique - 5} more unique values)")
```

# **Overall Completeness**

```python
total_cells = df.shape[0] * df.shape[1]
null_cells = df.isnull().sum().sum()
completeness = 1 - null_cells / total_cells


print(f"  Completeness : {completeness:.1%}")
print(f"  Columns      : {df.shape[1]}")
print(f"  Rows         : {df.shape[0]:,}")
```
